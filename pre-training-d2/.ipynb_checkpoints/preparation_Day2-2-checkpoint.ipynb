## 1. Data Visualization with Matplotlib & Seaborn

### Apa Itu Matplotlib dan Seaborn?

__Matplotlib__ adalah library visualisasi data paling dasar dan fleksibel dalam Python. Dengan Matplotlib, Anda bisa membuat berbagai jenis grafik, seperti garis, batang, scatter plot, histogram, dan lainnya. Karena fleksibilitasnya, Matplotlib sering digunakan untuk membuat grafik kustom.

__Seaborn__ adalah library berbasis Matplotlib yang dirancang untuk mempermudah pembuatan grafik statistik. Library ini menyediakan fungsi yang sederhana untuk membuat visualisasi seperti heatmap, box plot, pair plot, dan lainnya, dengan tampilan yang lebih estetis dibandingkan Matplotlib.

### Mengapa Visualisasi Data Penting?

Visualisasi data memiliki beberapa manfaat penting:

1. __Mengidentifikasi Pola dan Tren__: Grafik mempermudah analisis data yang kompleks.
2. __Mendeteksi Outlier__: Data yang tidak sesuai pola dapat dengan cepat dikenali melalui grafik.
3. __Komunikasi Informasi__: Visualisasi membantu menyampaikan insight kepada audiens dengan cara yang mudah dipahami.

### Langkah Awal Menggunakan Matplotlib

Matplotlib digunakan melalui modul utamanya, pyplot. Berikut adalah cara dasar membuat grafik sederhana dengan Matplotlib:

### Membuat Grafik Garis

````python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Membuat Grafik
plt.plot(x, y, marker='o', color='blue', label='Garis 1')
plt.title("Contoh Grafik Garis")
plt.xlabel("Sumbu X")
plt.ylabel("Sumbu Y")
# plt.legend()
plt.grid()
plt.show()
````

### Membuat Grafik Batang

````python
# Data
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]

# Membuat Grafik Batang
plt.bar(categories, values, color='blue')
plt.title("Contoh Grafik Batang")
plt.xlabel("Kategori")
plt.ylabel("Nilai")
plt.show()
````

### Membuat Histogram
Histogram digunakan untuk menampilkan distribusi data.

````python
import numpy as np

# Data
data = np.random.randn(1000)

# Membuat Histogram
plt.hist(data, bins=20, color='blue', edgecolor='black')
plt.title("Distribusi Data")
plt.xlabel("Nilai")
plt.ylabel("Frekuensi")
plt.show()
````

### Visualisasi Data dengan Seaborn

Seaborn menyederhanakan proses pembuatan visualisasi dan menambahkan elemen statistik yang berguna. Berikut adalah contoh penggunaan Seaborn:

### Membuat Scatter Plot

Scatter plot digunakan untuk menunjukkan hubungan antara dua variabel.

````python
import seaborn as sns

# Data
import pandas as pd
data = pd.DataFrame({
    'X': [1, 2, 3, 4, 5],
    'Y': [5, 10, 15, 20, 25],
    'Category': ['A', 'B', 'A', 'B', 'A']
})

# Scatter Plot
sns.scatterplot(x='X', y='Y', hue='Category', data=data, palette='deep')
plt.title("Contoh Scatter Plot")
plt.show()
````

### Membuat Box Plot

Box plot digunakan untuk melihat distribusi data dan mendeteksi outlier.

````python
# Data
data = pd.DataFrame({
    'Category': ['A', 'A', 'B', 'B'],
    'Value': [7, 8, 6, 9]
})

# Box Plot
sns.boxplot(x='Category', y='Value', data=data)
plt.title("Contoh Box Plot")
plt.show()
````

### Membuat Heatmap
Heatmap digunakan untuk menampilkan matriks data dalam format visual.

````python
# Data
matrix = np.random.rand(5, 5)

# Heatmap
sns.heatmap(matrix, annot=True, cmap='coolwarm')
plt.title("Contoh Heatmap")
plt.show()
````

## 📊 Exploratory Data Analysis (EDA) dengan Python

__Apa itu EDA?__

Proses eksplorasi awal data sebelum analisis lanjutan atau pemodelan.

dengan Tujuan:

* Memahami struktur dan karakteristik data.
* Menemukan pola, outlier, atau anomali.
* Menyusun hipotesis awal.

### 📈 Import Library


````python
# Import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Konfigurasi visualisasi
sns.set(style="whitegrid")
````

### 🗃️ Load Dataset

📋 __Deskripsi Dataset__ ```tips```

Dataset ini berisi catatan transaksi tip yang diberikan pelanggan di sebuah restoran.

✅ __Konteks__:

* Data dikumpulkan oleh seorang pelayan pria di sebuah restoran Amerika sekitar tahun 1990-an.
* Pelayan ini mencatat berapa besar tagihan pelanggan, berapa tip yang mereka beri, hari apa, jam berapa, dan siapa yang membayar.

🔖 __Fitur-fitur dalam Dataset__ ```tips```

| Kolom       | Tipe Data | Deskripsi                                           |
|-------------|-----------|-----------------------------------------------------|
| total_bill  | float     | Total tagihan pelanggan (dalam dolar Amerika)       |
| tip         | float     | Tip yang diberikan pelanggan (dalam dolar Amerika)  |
| sex         | category  | Jenis kelamin pelanggan (`Male` / `Female`)        |
| smoker      | category  | Apakah pelanggan seorang perokok (`Yes` / `No`)    |
| day         | category  | Hari kunjungan (`Thur`, `Fri`, `Sat`, `Sun`)       |
| time        | category  | Waktu makan (`Lunch` / `Dinner`)                   |
| size        | int       | Jumlah orang dalam rombongan pelanggan             |

### 📈 Univariate Analysis (Analisis Satu Variabel)

Univariate analysis adalah jenis analisis data yang hanya melibatkan satu variabel saja. Tujuan dari analisis ini adalah untuk memahami karakteristik dasar dari variabel tersebut, seperti distribusi data, nilai rata-rata, median, minimum, maksimum, dan sebaran nilai. Univariate analysis sering digunakan untuk mendapatkan gambaran umum mengenai data sebelum melakukan analisis lebih lanjut.

````python
# Distribusi Data Numerik

# Histogram
plt.hist(df['total_bill'], bins=20, color='skyblue')
plt.title('Distribusi Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Frekuensi')
plt.show()

# Boxplot
sns.boxplot(y=df['total_bill'])
plt.title('Boxplot Total Bill')
plt.show()
````

````python
# Distribusi data kategorik

sns.countplot(x='day', data=df)
plt.title('Frekuensi Tiap Hari')
plt.show()
````

### 📉 Bivariate Analysis (Analisis Dua Variabel)

Bivariate analysis adalah analisis yang melibatkan dua variabel sekaligus untuk mengetahui hubungan atau asosiasi antara keduanya. Tujuan dari bivariate analysis adalah untuk melihat apakah ada pola, korelasi, atau pengaruh antara satu variabel dengan variabel lainnya.

````python
# Hubungan dua data numerik

sns.scatterplot(x='total_bill', y='tip', data=df)
plt.title('Total Bill vs Tip')
plt.show()

# Korelasi
df[['total_bill', 'tip']].corr()
````

````python
# Hubungan Variabel Numerik dan Kategorik

sns.boxplot(x='sex', y='total_bill', data=df)
plt.title('Total Bill berdasarkan Jenis Kelamin')
plt.show()
````

### 📊 Multivariate Analysis (Analisis Lebih dari Dua Variabel)

Multivariate analysis adalah analisis yang melibatkan tiga atau lebih variabel dalam satu waktu. Tujuan dari analisis ini adalah untuk memahami hubungan yang lebih kompleks antar variabel serta bagaimana interaksi di antara mereka mempengaruhi hasil akhir.

````python
# Pairplot

sns.pairplot(df)
plt.show()
````

````python
# Heatmap korelasi

# Ambil hanya kolom numerik
corr = df.select_dtypes(include=['float64', 'int64']).corr()

# Plot heatmap-nya
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Heatmap Korelasi')
plt.show()
````

### 🏴󠁡󠁺󠁳󠁵󠁳󠁿 Mendeteksi Outlier

````python
# Boxplot

sns.boxplot(x=df['total_bill'])
plt.show()
````

````python
# Z-score

from scipy import stats

z_scores = np.abs(stats.zscore(df['total_bill']))
df_outliers = df[(z_scores > 3)]

df_outliers
````

📌 __Penjelasan__:

Z-Score berguna untuk mengukur seberapa jauh data dari rata-rata (dalam satuan standar deviasi).

Nilai z > 3 biasanya dianggap outlier.

### 🌞 Insight dari EDA

Setelah eksplorasi, beberapa insight yang bisa didapat :

1. Total bill tertinggi terjadi di akhir pekan.
2. Pelanggan pria cenderung memberikan tip lebih besar.
3. Ada outlier di total_bill yang perlu diproses sebelum modeling.
