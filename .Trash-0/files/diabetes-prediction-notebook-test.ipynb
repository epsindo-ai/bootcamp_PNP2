<div style="text-align: justify;">

Berdasarkan informasi dari sumbernya, kita dapat mengetahui bahwa kumpulan data tersebut mencakup 9 kolom atribut, antara lain:


1. **gender**: Gender mengacu pada jenis kelamin biologis seseorang, yang dapat berdampak pada kemungkinan mereka terkena diabetes. Ada tiga tipe di dalamnya (pria, wanita, lainnya)


2. **age**: Usia merupakan faktor penting karena diabetes sering didiagnosis pada orang tua. Usia berkisar antara (0-80) dalam kumpulan data kami.
3. **hypertension**: Hipertensi adalah suatu kondisi medis di mana tekanan darah di arteri terus meningkat. Memiliki nilai (0 atau 1) dimana 0 menunjukkan tidak menderita hipertensi dan 1 berarti menderita hipertensi.
4. **heart_disease**: Penyakit jantung adalah kondisi medis lain yang terkait dengan peningkatan risiko terkena diabetes. Memiliki nilai (0 atau 1) dimana 0 menunjukkan tidak mengidap penyakit jantung dan 1 berarti memang mengidap penyakit jantung.
5. **smoking_history**: Riwayat merokok juga dianggap sebagai faktor risiko diabetes dan dapat memperburuk komplikasi terkait diabetes. Dalam kumpulan data kami, kami memiliki 5 kategori (bukan saat ini, sebelumnya, Tidak Ada Info, saat ini, tidak pernah dan selamanya)
6. **BMI**: adalah ukuran lemak tubuh berdasarkan berat dan tinggi badan. Nilai BMI yang lebih tinggi dikaitkan dengan risiko diabetes yang lebih tinggi. Kisaran BMI dalam kumpulan data adalah dari (10,16 hingga 71,55). BMI di bawah 18,5 berarti kurus, 18,5-24,9 normal, 25-29,9 kelebihan berat badan, dan 30 atau lebih berarti obesitas.
7. **HbA1c_level**: merupakan ukuran rata-rata kadar gula darah seseorang selama 2-3 bulan terakhir. Kadar yang lebih tinggi menunjukkan risiko lebih tinggi terkena diabetes. Sebagian besar kadar HbA1c di atas 6,5% mengindikasikan diabetes.
8. **blood_glucose_level**: Kadar glukosa darah mengacu pada jumlah glukosa dalam darah pada waktu tertentu. Kadar gula darah yang tinggi merupakan tanda utama penyakit diabetes.
9. **Diabetes**: Merupakan variabel target dengan nilai prediksi (0 atau 1) dengan nilai 1 mewakili adanya diabetes dan 0 mewakili risiko penyakit.
</div>

## Model inference demo

Setelah model sudah dibangun, selanjutnya adalah menyimpan model ke dalam Pickle, dengan nama 'model.pkl'.

```python
import pickle 

pickle.dump(rfc, open('model.pkl', 'wb'))
```

Kemudian load model, dan mencoba untuk melakukan prediksi dengan input data berupa array yang dikonversi menjadi numpy array yang mengubah array asli menjadi array 2D dengan satu baris dan sebanyak mungkin kolom sesuai ukuran total array asli.

```python
input_data = (0,36.0,0,0,1,23.45,5.0,155)
# changing the input_data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the array as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

model = pickle.load(open('model.pkl', 'rb'))

prediction = model.predict(input_data_reshaped)
print(prediction)

if (prediction[0] == 0):
  print('The person is not diabetic')
else:
  print('The person is diabetic')
```
