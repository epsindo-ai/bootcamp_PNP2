## Data Cleaning & Preparation


Melakukan pemeriksaan pada data yang memiliki nilai yang kembar (duplikat) dan menghapusnya, lalu melihat bentuk data (jumlah baris, jumlah kolom) setelah dihapus nilai yang kembar.

```python
df.drop_duplicates(inplace=True)
df.shape
```

Menampilkan banyaknya nilai dari kolom 'diabetes', untuk melihat apakah data pada kolom tersebut tidak seimbang.

```python
df['diabetes'].value_counts()
```

Terlihat kolom 'diabetes' yang akan menjadi target prediksi mengalami data yang tidak seimbang, dimana jumlah data yang tidak terkena penyakit Diabetes jauh lebih banyak dari data yang terkena penyakit tersebut. 
\
\
Untuk menangani masalah ini, kita dapat menggunakan SMOTE yaitu teknik untuk augmentasi data yang dirancang untuk mengurangi masalah ketidakseimbangan data pada data 'diabetes'.

Melakukan konversi pada nilai kategori menjadi nilai numerik, ini digunakan untuk membuat matriks korelasi.

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
from pandas.core.dtypes.common import is_numeric_dtype
for column in df.columns:
    if is_numeric_dtype(df[column]):
        continue
    else:
        df[column]=label.fit_transform(df[column])
```

Melihat 5 data teratas dari dataframe yang sudah diubah menjadi nilai numerik.

```python
df.head()
```

Menampilkan matriks korelasi dari data.

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(),annot = True )
plt_title = plt.title("Correlation Matrix")
```

Kesimpulan matriks korelasi:
* Kolom HbA1c_level dan kolom blood_glucose_level memiliki korelasi positif terhadap diabetes

Melakukan penanganan pada data yang tidak seimbang dengan SMOTE.

```python
from imblearn.over_sampling import SMOTE

X = df.drop('diabetes', axis=1)
Y = df['diabetes']

# Y.value_counts()
# X.unique()

smk = SMOTE(random_state=42)
X_resampled, Y_resampled = smk.fit_resample(X, Y)

resampled_df = pd.concat([pd.DataFrame(X_resampled, columns=X.columns),
                          pd.DataFrame(Y_resampled, columns=['diabetes'])], axis=1)
resampled_df.head()
```

Menampilkan banyaknya nilai dari kolom 'diabetes', yang telah di resampling dengan SMOTE. Hasil nilai menjadi seimbang.

```python
resampled_df['diabetes'].value_counts()
```

## Build Machine Learning model

Membagi dataframe menjadi dua bagian, yaitu data untuk melatih model dengan perbandingan 80% dari dataframe  dan data untuk menguji model dengan perbandingan 20% dari dataframe.

```python
from sklearn.model_selection import train_test_split
X = resampled_df.drop('diabetes', axis=1)
Y = resampled_df['diabetes']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
print(X_train.shape)
print(X_test.shape)
```

Membangun model Random Forest untuk melakukan prediksi penyakit Diabetes.

```python
from sklearn.ensemble import RandomForestClassifier 
rfc = RandomForestClassifier()
rfc.fit(X_train, Y_train)
accuracy = rfc.score(X_test, Y_test)
print("Accuracy:", accuracy)
```

Hasil akurasi dari model sebesar 97 %

Selanjutnya, mengevaluasi model yang telah dibangun, melihat hasil presisi, recall, dan f1-score.

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
rfc_predict=rfc.predict(X_test)
rfc_predict
cr=classification_report(Y_test,rfc_predict)
print (cr)
```

Menampilkan hasil prediksi dari data uji dengan menggunakan confusion matriks.

```python
cm = confusion_matrix(Y_test, rfc_predict)
sns.heatmap(cm,annot=True)
plt.xlabel("predict value")
plt.ylabel('Actual value')
```

## Model inference demo

Setelah model sudah dibangun, selanjutnya adalah menyimpan model ke dalam Pickle, dengan nama 'model.pkl'.

```python
import pickle 

pickle.dump(rfc, open('model.pkl', 'wb'))
```

Kemudian load model, dan mencoba untuk melakukan prediksi dengan input data berupa array yang dikonversi menjadi numpy array yang mengubah array asli menjadi array 2D dengan satu baris dan sebanyak mungkin kolom sesuai ukuran total array asli.

```python
input_data = (0,36.0,0,0,1,23.45,5.0,155)
# changing the input_data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the array as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

model = pickle.load(open('model.pkl', 'rb'))

prediction = model.predict(input_data_reshaped)
print(prediction)

if (prediction[0] == 0):
  print('The person is not diabetic')
else:
  print('The person is diabetic')
```
