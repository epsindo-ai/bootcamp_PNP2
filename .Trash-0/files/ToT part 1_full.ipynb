### STRING FORMATTING

```python
name = 'bayu'
age = 17

# Old Style

print('Hello i am %s i am %d years old' % (name, age))

# 'New Style' string formatting

print('Hello i am {0} i am {1} years old'.format(name, age))

# String interpolation

print(f'Hello i am {name} i am {age} years old')
```

### Method
---
Sederhananya, method adalah suatu fungsi yang dimiliki oleh suatu object.

Apa itu object? Segala yang ada di python adalah object. Contohnya kita membuat variabel bertipe string, bertipe list, bertipe numeric dan lain sebagainya, itu semua adalah object, dan setiap object memiki fungsi yang hanya bisa digunakan oleh object tersebut. Sebagai contoh, object string memiliki method uppercase. Jadi intinya, fungsi yang dimiliki oleh suatu object dinamakan method

```python
# membuat object string
huruf_kecil = 'huruf_kecil'
# memanggil salah satu method string
print(huruf_kecil.upper())
```

```python
huruf_besar = 'GUNADARMA'
print(huruf_besar.lower())
```

```python
#mendifinisikan object list
keluarga_metal = ['ayah', 'ibu', 'si metal', 'kakak']

# memanggil method yang dimiliki oleh list
print(keluarga_metal.index('si metal'))
```

### List
---
List adalah salah satu tipe data untuk mengoleksikan data di python. Contohnya, kita punya koleksi data suhu semua anggota keluarga kita, untuk menyimpan data ini, tentunya lebih mudah kalau kita menyimpannya dalam satu tempat, dan tempat itu salah satunya bisa bernama list. Mengapa salah satunya? Karena tempat menyimpan koleksi data di python tidak hanya list, ada tipe data lain seperti dictionary, tuple, set dan lain-lain. Namun dalam pembahasan ini kita akan fokus membahas tentang list.

```python
#Membuat data list

berat_badan = [
                62, # index 0 
                77, # index 1
                82, # index 2
                50, # index 3
                66 # index 4
                ]

# Subsetting list
print(berat_badan[0]) # posisi pertama

print(berat_badan[-2]) # posis kedua dari belakang

# Slicinglist
print(berat_badan[:4]) # Mengambil data dengan index 0, 1, 2, 3

print(berat_badan[2:4]) 
```

### Manipulasi List

```python
# merubah nilai list 
berat_badan[3] = 22
berat_badan
```

```python
berat_badan[-2:] = [1,2]
berat_badan
```

Dalam menambahkan elemen dalam list, kita bisa menggunakan metode .append() ataupun .insert() seperti contoh berikut:

```python
murid = ['kevin', 'reja', 'bagus', 'alfinka', 'bayu']
murid.append('amel')
murid
```

```python
murid.insert(2, 'khansa')
murid
```

```python
murid.pop(2)
murid
```

### Dictionary
---
Dictionary adalah suatu topik yang sangat penting dalam python dan data science. Karena dictionary seperti penyusun untuk suatu objek yang lebih kompleks seperti pandas dataframe yang akan kita pelajari nanti.

Jadi, dictionary ini adalah salah satu jenis tipe data di python yang memetakan antara key dan value dari suatu data

```python
dict = {'one':1, 'two':2, 'three':3}
dict
```

```python
dict.values()
```

```python
dict.keys()
```

#### Akses Anggota Dictionary
---
Dictionary tidak menggunakan indeks. Anggota dictionary diakses dengan menggunakan kuncinya

```python
# Kita dapat mengakses suatu nilai dalam dictionary dengan cara seperti ini:
print(dict['one'])
```

```python
#Dengan menggunakan fungsi get(), bila kunci tidak ada di dalam dictionary, maka akan dikembalikan None
print(dict.get('five'))
```

#### Menambahkan & Menghapus Dictionary
---
Kita bisa menghapus anggota tertentu pada dictionary dengan menggunakan fungsi pop(). Fungsi ini menghapus anggota dengan mengembalikan kunci dari anggota tersebut.

Fungsi lain, popitem() digunakan untuk menghapus anggota acak dari dictionary. Untuk menghapus semua anggota dictionary, bisa menggunakan fungsi clear().

```python
# Menambah entri baru
dict['four'] = 4
dict
```

```python
# hapus hapus entri dengan func pop()
dict.pop('four')
dict
```

```python
# hapus entri dengan func del
del dict['two']
dict
```

```python
#dict.clear()
```

#### Mengubah Anggota Dictionary
---
Dictionary bersifat mutable. Kita bisa menambahkan atau mengubah nilai dari anggotanya menggunakan operator penugasan. Bila kunci sudah ada, maka nilainya yang akan diupdate. Bila kunci belum ada, maka akan ditambahkan sebagai kunci baru.

```python
#Kita juga dapat mengubah suatu nilai dalam dictionary dengan cara berikut:

dict['one'] = 'satu'
print(dict)
```

```python
#loop terhadap key pada dictionary
for k in dict:
    print("{} = {}".format(k, dict[k]))
    
print('')
#loop terhadap value / nilai nya saja pada dictionary    
for k in dict.values():
    print("{}\n".format(k))   

    
print('')

for y in dict.keys():
    print('{}'.format(y))
```
